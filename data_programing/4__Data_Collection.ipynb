```python
import requests
import bs4
import pandas as pd
 
kw = '공룡'
url = f'https://news.google.com/rss/search?q={kw}&hl=ko&gl=KR&ceid=KR%3Ako'

data = requests.get(url).text
data

```

```python
news_data = bs4.BeautifulSoup(data, features = 'xml')
# news_data = bs4.BeautifulSoup(data, 'lxml')
news_data
```

```python
news_list = news_data.find_all('item') 
news_list
```

```python
all_news_list = []
for news in news_list:
    title = news.find('title').text
    source = news.find('source').text
    one_news_dict = {'title': title, 'source': source}
    all_news_list.append(one_news_dict)
pd.DataFrame(all_news_list)
```

```python
date = '2020-08-01'
url = 'https://api.sunrise-sunset.org/json?lat=37.5&lng=127.0&date='
data = requests.get(url + date).json()
data['results']
```

```python
def by_date(date):
    url = 'https://api.sunrise-sunset.org/json?lat=37.5&lng=127.0&date='
    data = requests.get(url + date).json()
    return data['results']
```

```python
by_date('2020-02-01') # example of the usage of function
```

```python
import pandas as pd
sample_list = []
sample_list.append(by_date('2020-01-01'))
sample_list.append(by_date('2020-01-02'))
sample_list.append(by_date('2020-01-03'))
df = pd.DataFrame(sample_list)
df
```

```python
import requests
import xmltodict
url= 'https://news.google.com/rss/search?q=data%20science&hl=ko&gl=KR&ceid=KR%3Ako'
```

```python
data = requests.get(url).text
news_data = xmltodict.parse(data)
news_data
```

```python
import xml.dom.minidom
dom = xml.dom.minidom.parseString(data)
pretty_xml = dom.toprettyxml()
print(pretty_xml)
```

```python
news_list = news_data['rss']['channel']['item']
for num in range(0, len(news_list)):
    print( news_list[num]['title'] )
```

```python
import pandas as pd
news_for_dic = []
for num in range(0, len(news_list)):
    news_for_dic.append( { news_list[num]['title']: news_list[num]['source']['#text']} )
dic = { 'Title': [news_for_dic[i].keys() for i in range(0, len(news_for_dic))], 
    'Source': [news_for_dic[i].values() for i in range(0, len(news_for_dic))]  }
news_df = pd.DataFrame(dic)
news_df.head()
```

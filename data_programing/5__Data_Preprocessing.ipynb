```python
from imblearn.over_sampling import SMOTE
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from collections import Counter
```

```python
# Create a synthetic imbalanced classification dataset
X, y = make_classification(n_classes=2, class_sep=2, flip_y=0,
                           weights=[0.1, 0.9], n_features=20, n_clusters_per_class=1, n_samples=1000)
print(f"Before SMOTE: {Counter(y)}")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
sm = SMOTE(random_state=42)
X_res, y_res = sm.fit_resample(X_train, y_train)
print(f"After SMOTE: {Counter(y_res)}")
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X_train[y_train == 0][:, 0], X_train[y_train == 0][:, 1], label="Class #0", alpha=0.5, linewidth=0.15)
plt.scatter(X_train[y_train == 1][:, 0], X_train[y_train == 1][:, 1], label="Class #1", alpha=0.5, color='r', linewidth=0.15)
plt.title('Original class distribution')
plt.legend()
plt.subplot(1, 2, 2)
plt.scatter(X_res[y_res == 0][:, 0], X_res[y_res == 0][:, 1], label="Class #0", alpha=0.5, linewidth=0.15)
plt.scatter(X_res[y_res == 1][:, 0], X_res[y_res == 1][:, 1], label="Class #1", alpha=0.5, color='r', linewidth=0.15)
plt.title('After SMOTE class distribution')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# 예제 데이터프레임을 생성합니다.
df = pd.DataFrame({
    'Color': ['Red', 'Green', 'Blue', 'Green', 'Red'],
    'Size': ['S', 'M', 'L', 'S', 'M'],
    'Price': [10, 15, 20, 15, 10]
})
```

```python
df
```

```python
df_encoded = pd.get_dummies(df, columns=['Color', 'Size'])
df_encoded
```

```python
df_encoded = pd.get_dummies(df, columns=['Color', 'Size'], dtype=int)
df_encoded
```

```python
# 필요한 라이브러리를 임포트합니다.
from sklearn.preprocessing import OneHotEncoder


# OneHotEncoder 객체를 생성합니다.
encoder = OneHotEncoder(sparse_output=False)
# 범주형 변수에 대해서만 one-hot encoding을 수행합니다.
# fit_transform 메서드는 두 단계를 한번에 수행합니다.
encoded_data = encoder.fit_transform(df[['Color', 'Size']])
encoded_data
```

```python
# One-hot encoding된 데이터를 DataFrame으로 변환합니다.
# encoder.get_feature_names_out() 메서드를 사용하여 열 이름을 얻습니다.
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(['Color', 'Size']))

# 원래의 'Price' 열을 인코딩된 데이터프레임에 추가합니다.
encoded_df['Price'] = df['Price']

encoded_df
```

```python
# Color와 Size 열에 대해 Label Encoding 수행
df['Color_encoded'] = pd.factorize(df['Color'])[0]
df['Size_encoded'] = pd.factorize(df['Size'])[0]
df
```

```python
from sklearn.preprocessing import LabelEncoder
le_color = LabelEncoder()
le_size = LabelEncoder()
df['Color_encoded'] = le_color.fit_transform(df['Color'])
df['Size_encoded'] = le_size.fit_transform(df['Size'])
```

```python
import pandas as pd

# 샘플 데이터 생성
data = {
    'Education': ['High School', 'Bachelors', 'Masters', 'PhD', 'Associates', 'Bachelors', 'Masters'],
    'Income': ['Low', 'Low', 'Medium', 'High', 'Low', 'Medium', 'High']
}

df = pd.DataFrame(data)
df
```

```python
education_order = {
    'High School': 1,
    'Associates': 2,
    'Bachelors': 3,
    'Masters': 4,
    'PhD': 5
}

income_order = {
    'Low': 1,
    'Medium': 2,
    'High': 3
}

# 딕셔너리를 사용하여 Ordinal Encoding 수행
df['Education_encoded'] = df['Education'].map(education_order)
df['Income_encoded'] = df['Income'].map(income_order)
df
```

```python
from sklearn.preprocessing import OrdinalEncoder

# 범주의 순서를 정의
education_categories = ['High School', 'Associates', 'Bachelors', 'Masters', 'PhD']
income_categories = ['Low', 'Medium', 'High']

# OrdinalEncoder 인스턴스 생성
encoder = OrdinalEncoder(categories=[education_categories, income_categories])
# 인코딩할 열을 선택하고 fit_transform 메소드를 적용
df_encoded = encoder.fit_transform(df[['Education', 'Income']])

# 인코딩된 데이터를 새로운 열에 할당
df['Education_encoded'] = df_encoded[:, 0]
df['Income_encoded'] = df_encoded[:, 1]

df
```
